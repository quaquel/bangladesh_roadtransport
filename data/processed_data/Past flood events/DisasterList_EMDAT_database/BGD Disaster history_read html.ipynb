```python
import pandas as pd
import xlrd 
from xlrd import open_workbook
from bs4 import BeautifulSoup
import lxml
import html5lib
```

```python
directory = 'D:/sibeleker/surfdrive/Documents/Bangladesh/DisasterList/' 
cols = ['Start date', 'End date', 'Country name', 'ISO', 'Location', 'Disaster type', 'Disaster subtype', 'Associated dis.',\
          'Associated dis2.', 'Total deaths', 'Total affected', 'Total damage', 'Insured loss', 'Disaster ID'] 
df = pd.DataFrame(columns=cols)
for i in range(14):
    df_t = pd.read_html(directory+'disaster_list_{}.html'.format(i+1), header=0)
    for d in df_t[12:]:
        s = pd.Series(list(d.columns), index=cols)
        df = df.append(s, ignore_index=True)
        df = df.drop_duplicates()
df
```

```python
a = ['--.1', '--.2', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11','Unnamed: 12']
b = ['--']
df = df.replace(a,'--')
```

```python
df.to_csv(directory+'DisasterList.csv')
df.to_excel(directory+'DisasterList.xlsx')
```
